```python
# input data rumah sakit
input_data_rs = Input(shape=(4,))

# input data user
input_data_us = Input(shape=(6,))

# Layer untuk data rumah sakit dan user
concatenated = Concatenate()([input_data_rs, input_data_us])

# Hidden Layer data rumah sakit dan user
hidden_layer_1 = Dense(64, activation='relu')(concatenated)
hidden_layer_2 = Dense(128, activation='relu')(hidden_layer_1)
hidden_layer_3 = Dense(32, activation= 'relu')(hidden_layer_2)

# Output Layer
output_layer = Dense(10, activation='linear')(hidden_layer_3)

# Model
model = Model(inputs=[input_data_rs, input_data_us], outputs=output_layer)

# Optimizer
model.compile(optimizer='adam', loss='mse')

model.summary()

```
